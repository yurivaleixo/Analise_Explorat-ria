```python
df_idade_filtro = df[(df['IDADE'] >= 20) & (df['IDADE'] <= 60)]
```

```python
import plotly.express as px
grafico =  px.histogram (df_idade_filtro,  x = "IDADE", nbins=25, color='IDADE') 
grafico.update_layout(width=500,height=500,title_text='Solicitantes entre 20 até 60 anos',
                      title_font_color = 'black') 
grafico.show()
```
