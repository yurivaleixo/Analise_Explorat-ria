Teste Normalidade

```python
import plotly.express as px
grafico =  px.histogram (df_156,  x = "IDADE", nbins=32, histnorm='percent',) 
grafico.update_layout(width=1000,height=400,title_text='Distribuição das idades') 
grafico.show()
```

```python
import plotly.express as px
grafico =  px.histogram (desc_tupa,  x = "IDADE", nbins=32, histnorm='percent',) 
grafico.update_layout(width=1000,height=400,title_text='Distribuição das idades') 
grafico.show()
```

```python
import seaborn as sns
sns.histplot(df_156, x='IDADE', bins=32, color="pink", kde=True, stat="probability");
```

QQPLOT

```python
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
stats.probplot(df_156['IDADE'], dist="norm", plot=plt)
plt.title("Normal Q-Q plot")
plt.show()
```

```python
stats.probplot(desc_tupa['IDADE'], dist="norm", plot=plt)
plt.title("Normal Q-Q plot")
plt.show()
```
